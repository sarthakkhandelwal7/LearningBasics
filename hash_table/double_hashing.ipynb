```python
class DoubleHashing:
    def __init__(self,size):
        self.hash_table = ['']*size
        self.size = size
        self.nearest_prime = self.find_nearest(size)
        
    def find_nearest(self,size):
        nearest_prime = 0
        for i in range(self.size):
            check = 0
            for j in range(2,i):
                if i % j == 0:
                    check+=1
                    break
            if (check == 0 and i < self.size):
                nearest_prime = i
        return nearest_prime
    
    def hash1(self,key):
        return key % self.size
    
    def hash2(self,key):
        return self.nearest_prime - (key % self.nearest_prime)
    
    def insert(self,key):
        i = 0
        index = (self.hash1(key) + i * self.hash2(key)) % self.size
                     
        if type(self.hash_table[index]) is str:
            self.hash_table[index] = key
            
        else:
            while type(self.hash_table[index]) is not str:
                i+=1
                index = (self.hash1(key) + i*self.hash2(key)) % self.size
                
            
            self.hash_table[index] = key
    
    def delete(self,key):
        i = 0
        index = (self.hash1(key) + i*self.hash2(key)) % self.size
        while(self.hash_table[index] != ''):
            if(self.hash_table[index] == key):
                self.hash_table[index] = 'Del'
                break
            i+=1
            index = (self.hash1(key) + i*self.hash2(key)) % self.size
    
    def search(self,key):
        i = 0
        index = (self.hash1(key) + i*self.hash2(key)) % self.size
        while(self.hash_table[index] != ''):
            if(self.hash_table[index] == key):
                print("Present")
                break
            i+=1
            index = (self.hash1(key) + i*self.hash2(key)) % self.size
        if self.hash_table[index] == '':
            print("Not present")
    
    def __str__(self):
        st = ''
        for i in range(self.size):
            st += f'{i}'
            if self.hash_table[i] != '':
                st += '-->'+f'{self.hash_table[i]}'
            st += '\n'
        return st
```
