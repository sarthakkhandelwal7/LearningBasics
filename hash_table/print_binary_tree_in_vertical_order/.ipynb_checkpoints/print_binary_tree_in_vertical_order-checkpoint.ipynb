```python
from linked_list import LinkedList
class Node:
    def __init__(self,data):
        self.left = None
        self.right = None
        self.val = data

class BinaryTreeInVerticalOrder:
    def __init__(self,root):
        self.root = root
        self.left_distance = 0
        self.right_distance = 0
        self.size = self.find_width() + 1
        self.root_position = abs(self.left_distance)
        self.hash_table = ['']*self.size
    
    
    def find_distance_from_root(self,root,left_distance,right_distance,distance):
       
        if root:
            if self.left_distance > distance:
                self.left_distance = distance
                                 
            elif self.right_distance < distance:
                self.right_distance = distance
                                
            self.find_distance_from_root(root.left,left_distance,right_distance,distance-1)
            self.find_distance_from_root(root.right,left_distance,right_distance,distance+1)
    
    def find_width(self):
        self.find_distance_from_root(self.root,0,0,0)
        return self.right_distance - self.left_distance
    
    def _hash(self,key):
        return abs(key)
        
    def get_vertical_order(self,root = None, position = None):
        if root == None and position == None:
            root = self.root
            position = self.root_position
            
        if root:
            index = self._hash(position)
            
            if type(self.hash_table[index]) is not str:
                self.hash_table[index].insert_node('e',root.val)
            
            else:
                self.hash_table[index] = LinkedList(0)
                self.hash_table[index].insert_node('h',root.val)
                
            self.get_vertical_order(root.left,position-1)
            self.get_vertical_order(root.right,position+1)
    
    def print_vertical_order(self):
        self.get_vertical_order()
        st = ''
        subst = ''
        for i in range(self.size):
            if type(self.hash_table[i]) is not str:
                curr = self.hash_table[i].head_node
                st+=''
                while curr != 'Null':
                    subst+=f'{curr.data} '
                    curr = curr.next
                st+=(subst[:len(subst)-1]+'\n')
                subst = ''
            else:
                st+='\n'
        print(st)
```

```python
root = Node(1) 
root.left = Node(2) 
root.right = Node(3) 
root.right.left = Node(5) 
root.right.right = Node(6) 
root.right.left.left = Node(7)
root.right.left.right = Node(8) 
root.right.left.right.left = Node(9)
root.right.left.right.right = Node(10)
b = BinaryTreeInVerticalOrder(root)
b.print_vertical_order()
```
